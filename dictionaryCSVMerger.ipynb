```python
len(scoreFiles)
```

```python
# create merged csv
for score in scoreFiles:
    print('------------'+score+'----------\n\n')
    with open(filePathToWeightedScores +"OpenRarityScores/"+score+'_openRarityScores.csv', 'r') as file:
        openRarityDF = pd.read_csv(file)
        openRarityDF.rename({'Rarity score' : 'OpenRarity score'}, axis=1, inplace=True)
        
    with open(filePathToWeightedScores +'TournamentScores/'+score+'_tournamentScores.csv', 'r') as file:
        tournamentDF=pd.read_csv(file)
        tournamentDF.rename({'Rarity score' : 'Tournament score'}, axis=1, inplace=True)
        
    with open(filePathToWeightedScores +'RarityToolsScores/'+score+'_rarityToolsScores.csv', 'r') as file:
        rarityToolsDF=pd.read_csv(file)
        rarityToolsDF.rename({'Rarity score' : 'RarityTools score'}, axis=1, inplace=True)
    
    if len(openRarityDF) == len(tournamentDF) and len(openRarityDF) == len(rarityToolsDF):
        mergedDF = pd.merge(openRarityDF, tournamentDF, on='Token Id')
        mergedDF = pd.merge(mergedDF, rarityToolsDF, on='Token Id')
        mergedDF = mergedDF[["Token Id", "OpenRarity score", "Tournament score", "RarityTools score"]]
    else:
        print('different len of rarity functions')
        print('openrarity -- ' + str(len(openRarityDF)))
        print('raritytools -- '+str(len(rarityToolsDF)))
        print('tournament -- '+str(len(openRarityDF)))
        
    if len(mergedDF) == len(openRarityDF):
        mergedDF.to_csv(filePathToWeightedScores+"Merged/"+score+'.csv', index=False)
    else:
        print('Потеря пакетов')
```
